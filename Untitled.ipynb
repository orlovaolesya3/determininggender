```python
from ultralytics import YOLO
import cv2

# Загрузка обученной модели YOLOv8
# Замените 'best.pt' на путь к вашей обученной модели
model = YOLO("best.pt")

# Функция для определения пола
def get_gender(cls):
    if cls == 0:  # Мужчина
        return "M"
    elif cls == 1:  # Женщина
        return "W"
    else:
        return "Неизвестно"

# Загрузка изображения
image_path = "foto.jpg"  # Укажите путь к вашему изображению
image = cv2.imread(image_path)

# Предсказание на изображении
results = model.predict(image, iou=0.2)

# Обработка результатов
for result in results:
    boxes = result.boxes  # Bounding boxes
    for box in boxes:
        # Координаты bounding box
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        
        # Класс и уверенность
        cls = int(box.cls[0])  # Класс (0 - мужчина, 1 - женщина)
        conf = float(box.conf[0])  # Уверенность
        
        # Определение пола
        gender = get_gender(cls)
        
        # Рисуем прямоугольник
        color = (0, 255, 0) if gender == "M" else (0, 0, 255)  # Зеленый для мужчин, красный для женщин
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
        
        # Добавляем текст
        label = f"{gender} {conf:.2f}"
        cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)

# Вывод изображения
cv2.imshow("YOLOv8 Detection", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
